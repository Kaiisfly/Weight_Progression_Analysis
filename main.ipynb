```python
# Count and Fill Missing Values
df.isnull().sum()
```

```python
df = df.set_index('Date')
df
```

```python
# Drop Duplicatated Rows
df.duplicated().sum()
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
df.shape
```

```python
df = df.drop(columns=['Notes', 'Workout Notes'])
df
```

```python
#Visualization
df = df.reset_index(drop=True)  # Reset index to remove duplicate labels
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Weight', y='RPE', data=df, color='blue', alpha=0.7)
plt.title('Weight vs RPE')
plt.xlabel('Weight')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Exercise Name', y='RPE', data=df)
plt.title('Box Plot of RPE by Exercise Name')
plt.xlabel('Exercise Name')
plt.ylabel('RPE')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Workout Name', y='Reps', data=df, palette='viridis')
plt.title('Bar Plot of Reps by Workout Name')
plt.xlabel('Workout Name')
plt.ylabel('Reps')
plt.xticks(rotation=45, ha='right')
plt.show()
```
